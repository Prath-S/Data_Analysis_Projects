```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('hotel_bookings 2.csv')
```

```python
df.head(15)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['company', 'agent'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')
df.head(15)
```

```python
a = df.select_dtypes(include=['object'])
for col in a:
    print(col, ': ', a[col].unique())
    print('\n')
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='is_canceled', data=df)
```

```python
plt.figure(figsize=(16,8))
df['Reservation_month_year'] = df.reservation_status_date.dt.to_period('M')
g = sns.countplot(x='Reservation_month_year', hue='is_canceled', data=df, palette='viridis')
g.set_xticklabels(g.get_xticklabels(), rotation=45)
plt.legend(loc= 'upper left', labels = ['Not Canceled', 'Canceled'])
plt.show()
```

```python
sns.countplot(x='hotel', hue='is_canceled', data=df)
plt.legend(loc= 'upper left', labels = ['Not Canceled', 'Canceled'])
```

```python
resort = df[df['hotel'] == 'Resort Hotel']
resort['is_canceled'].value_counts(normalize=True)
```

```python
city = df[df['hotel'] == 'City Hotel']
city['is_canceled'].value_counts(normalize=True)
```

```python
city = city.groupby('reservation_status_date')[['adr']].mean()
resort = resort.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
plt.plot(city.index, city['adr'], label='City Hotel')
plt.plot(resort.index, resort['adr'], label='Resort Hotel')
plt.legend(fontsize=16)
```

```python
plt.figure(figsize=(12,7))
ax = sns.countplot(x=df.reservation_status_date.dt.month_name(), hue=df.is_canceled, palette='YlGnBu')
plt.legend(labels = ['Not Canceled', 'Canceled'])
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df.reservation_status_date.dt.day_name(), hue=df.is_canceled, palette='YlOrRd')
plt.legend(loc= 'upper right', labels = ['Not Canceled', 'Canceled'])
```

```python
plt.figure(figsize=(10, 7))
df_1 = df[df['is_canceled']== 1].groupby(df.reservation_status_date.dt.month)[['adr']].sum().reset_index()
sns.barplot(df_1.reservation_status_date, df_1.adr, palette='Wistia')
```

```python
canceled_data = df[df['is_canceled']==1]
```

```python
top_10_countries = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(10,10))
colors = sns.color_palette('bright')
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index, colors=colors)
plt.show()
```
